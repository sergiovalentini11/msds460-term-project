```python
### Analyse the results

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os


### Analyzing Simulation Results

# read the simulation data from the csv file
df = pd.read_csv('001. Simulation Data/restaurant_simulation.csv')


# show first element of the event log
print(df['Event Log'][0])

# convert df['Event Log'][0] to a list of tuples
event_log = eval(df['Event Log'][0])

# create a data frame from the event log
event_log_df = pd.DataFrame(event_log, columns = ['Case ID', 'Time', 'Activity'])

# create a new column for the simulation number simid
event_log_df['simid'] = 0

# move simid to the front of the data frame
cols = event_log_df.columns.tolist()
cols = cols[-1:] + cols[:-1]
event_log_df = event_log_df[cols]

# show the first 10 rows of the event log data frame
print(event_log_df.head(10))

# now repeat the process for the entire data frame df and create a new data frame
# event_log_df_all that contains all the event log data for all simulations

# create an empty list to hold the data frames
dfs = []

# iterate over the rows of the data frame df
for i in range(df.shape[0]):
    event_log = eval(df['Event Log'][i])
    event_log_df = pd.DataFrame(event_log, columns = ['Case ID', 'Time', 'Activity'])
    event_log_df['simid'] = i
    cols = event_log_df.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    event_log_df = event_log_df[cols]
    dfs.append(event_log_df)

# concatenate the list of data frames into a single data frame
event_log_df_all = pd.concat(dfs)

# break line and show the first 10 rows of the event log data frame
print("\n")
print("---------------------------------------------------------")
print("\n")
print("First 10 rows of the event log data frame")
print("\n")
print(event_log_df_all.head(10))

# break line and show the last 10 rows of the event log data frame
print("\n")
print("---------------------------------------------------------")
print("\n")
print("Last 10 rows of the event log data frame")
print("\n")
print(event_log_df_all.tail(10))
```
